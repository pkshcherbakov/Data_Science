```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 29)
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('Table.xlsx')

df.head()
```

```python
#Удалим пустой столбец
df.drop(columns=['Unnamed: 6'], axis=1, inplace=True)
```

```python
df.info()
```

```python
#Незаполненных данных нет.
#Преобразуем данные в выбранных колонках в числовые. С форматом немного не понял как сделать, поэтому сначала 
#засплитил, потом сджойнил. Так работает. 
cols = ['Loan issued','Earned interest','Unpaid,  full amount']
for elem in cols:
    df[elem] = df[elem].str.replace('₽', '', regex=True)
    df[elem] = df[elem].str.split()
    df[elem] = df[elem].str.join('').astype('int64')
```

```python
#Задание о сводной таблице немного не понял. Интерпретировал его так: нужна таблица, у которой индексы это Рейтинг,
#Колонки это комиссия, а в значении либо сумма займов, либо среднее по сумме займов
df_pivot1 = df.pivot_table('Loan issued', index='Рейтинг', columns ='Comission, %', fill_value=0, aggfunc=(np.sum))
df_pivot2 = df.pivot_table('Loan issued', index='Рейтинг', columns ='Comission, %', fill_value=0, aggfunc=(np.mean))
```

```python
df_pivot2
```

```python
#Честно переписал формулу из задания. Искомый столбец назвал Таргетом
df['Loss']=df['Unpaid,  full amount'] * df.EL
df['InvestorProfit'] = df['Earned interest'] - df['Loss']
df['Profit'] = df.InvestorProfit + df['Comission, %']
df['Target'] = df['Profit'] / df['Loan issued']
#Вспомогательные колонки выбрасысаем.
#Новой информации они не несут. Также выбросим все, что не относится к факторам займов (я имеено так понял задание)
df.drop(columns=['Loss','InvestorProfit','Profit', 'Loan issued', 'Рейтинг', 'Comission, %', 'Earned interest', 'Unpaid,  full amount', 'EL' ], axis=1, inplace=True)
df.head()
```

```python
#Перед тем как строить любую модель посмотрим на датасет. 
cats = list(df.columns.values)
cats.remove('Target')
print(len(cats))
```

```python
X = df.drop("Target", axis=1)
y = df.Target
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, make_scorer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import BayesianRidge, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import lazypredict
from lazypredict.Supervised import LazyRegressor
import time
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.247, random_state=42)
```

```python
#Всего у нас 532 образца. Предлагаю 132 отобрать для теста. А обучаться будем на оставшихся 400.
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Я не знаю насколько сложные модели обычно в подобных задачах используются. 
#Единственный раз я считал прогноз дефолта на Kaggle недавно для AMEX. Там ключ к успеху был блендинг
#различных бустинговых моделей.
#Попробуем в качестве бейзлайн-моделей поисследовать всякие разные линейные модельки попроще
svr = LinearSVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'MAE:{mae}, MSE:{mse}')
```

```python
knn = KNeighborsRegressor(n_neighbors = 3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'MAE:{mae}, MSE:{mse}')
```

```python
bay = BayesianRidge()
bay.fit(X_train, y_train)
y_pred = bay.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'MAE:{mae}, MSE:{mse}')
```

```python
sgd = SGDRegressor()
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'MAE:{mae}, MSE:{mse}')
```

```python
#С линейными моделями пожалуй закончим, наибольшая точность у метода опорных векторов
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
y_pred = dtr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'MAE:{mae}, MSE:{mse}')
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_pred = rfr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'MAE:{mae}, MSE:{mse}')
```

```python
#Чтобы дальше вручную не перебирать различные варианты, я использую библиотеку lazypredict.
regr = LazyRegressor(verbose=0,
                     ignore_warnings=True,
                     custom_metric=None,
                     random_state=42, regressors='all')

models, predictions = regr.fit(X_train, X_test, y_train, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#C -- коэффициент регуляризации, max_iter -- количество итераций, dual -- выбор алгоритма
#Строк у нас мало, поэтому для оценки точности предпочтительно использовать кросс-валидацию.
svr = LinearSVR()
#По умолчанию гридсёрч пытается максимизировать оценку, поэтому сделаем свою собственную функцию потерь
mmssee = make_scorer(mean_squared_error,greater_is_better=False)
parameters = {'C': [0.7, 0.8, 0.9, 1.0], 'max_iter':[750, 1000, 1500, 2000], 'dual':[True, False]}
GridSearchCV_rgr = GridSearchCV(svr, parameters, cv=4, scoring=mmssee)
```

```python
GridSearchCV_rgr.fit(X_train, y_train)
```

```python
#Посмотрим на лучшие параметры
GridSearchCV_rgr.best_params_
```

```python
svr = LinearSVR(C=0.9, dual=True, max_iter=750)
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f'Итоговая точность:\nMAE:{mae}, MSE:{mse}')
```
